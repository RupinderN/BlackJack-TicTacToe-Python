```python
def space_check1():
    
    while True:
        
        First_Player_Coordinate = input("Choose your next position: (1-9) ")

        if markers[int(First_Player_Coordinate)] == " ":
            markers[int(First_Player_Coordinate)] = "X"
            break
        
        else:
            print("This spot is already full! Please choose again.")
            
```

```python
def space_check2():
    
    while True:
        
        Second_Player_Coordinate = input("Choose your next position: (1-9) ")

        if markers[int(Second_Player_Coordinate)] == " ":
            markers[int(Second_Player_Coordinate)] = "O"
            break
        
        else:
            print("This spot is already full! Please choose again.")
            
```

```python
print("Welcome to Tic-Tac-Toe!")
player_1 = input("Player 1: Do you want to be X or O? ").upper()

if player_1 == "X":
    print("Player 1 will go first.")
    player_2 = "O"

else:
    print("Player 2 will go first.")
    player_2 = "X"

start_game = input("Are you ready to play? Enter Yes or No. ").lower()

if start_game == "no":
    start_game = input("How about now? (You better fucking say 'Yes' or I'll personally come to your house and murder your entire family) ").lower()

keep_playing = True
markers = [" "," "," "," "," "," "," "," "," "," "]


if start_game == "yes":
    while keep_playing:
        
        board(markers)
        
        space_check1()
        
        board(markers)
        
        if player_1 == "X":
            keep_playing = win_check1(markers, "1")
        else:
            keep_playing = win_check1(markers, "2")


        if keep_playing == False:
            break
            
        
        space_check2()
        
        
        if player_2 == "O":
            keep_playing = win_check2(markers, "2")
        else:
            keep_playing = win_check2(markers, "1")
        
        if keep_playing == False:
            break

print("Thank you for playing!")
        
```
