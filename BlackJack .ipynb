```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(rank + " of " + suit)
    
    def __str__(self):
        return str(self.deck)

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        if values[card.split()[0]] == 11:
            self.aces +=1
        self.value += values[card.split()[0]]
        return self.cards
    
    def adjust_for_ace(self):
        if self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    
    def __init__(self,total=500):
        self.total = total
        self.bet = 0
        
        while True:
            try:
                self.bet = int(input(f"What would you like to bet? You have {self.total} chips. "))
                if 0 < self.bet <= self.total:
                    break
                else:
                    print("You may not input a number greater than your total chips or something less than zero. Try again.")

            except:
                print("It seems that you have not provided a correct input. Try again.")
                  
    def win_bet(self):
        self.total += self.bet

    def lose_bet(self):
        self.total -= self.bet
```
